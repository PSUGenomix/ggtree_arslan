```r
library(ggtree)
library(ggplot2)
library(ape)
```

```r
data(chiroptera) #load chiroptera data
```

```r
class(chiroptera) # chiroptera is an object of class "phylo"
```

```r
chiroptera
```

```r
names(chiroptera) #what are the variables in object?
```

```r
chiroptera$tip.label #names of OTUs are in variable tip.label. this can be accessed using $ sign
```

```r
chiroptera$Nnode #Number of nodes in tree
```

```r
chiroptera$edge #entire branching pattern is contained in this. very abstract. not readable by humans!
```

```r
chi.data<-fortify(chiroptera) #fortunately. fortify extracts all this info for you into a dataframe. 
head(chi.data)
```

```r
ggtree(chiroptera) #basic function to draw tree from tree object
```

```r
ggtree(chi.data,layout="circular")+geom_tiplab(aes(angle=angle)) #circularize tree and add label
```

```r
ggtree(chi.data,layout="rectangular")+geom_tiplab() #draw rectangular tree
```

```r
gzoom(chiroptera,grep('Rhinolophus',chiroptera$tip.label)) #zoom in on one taxonimic unit
```

```r
gzoom(chiroptera,grep('Hipposideros',chiroptera$tip.label)) # zoom in on another one
```

```r
split(chiroptera$tip.label,gsub("_\\w+", "", chiroptera$tip.label)) #groupin OTUs into broader categories
```

```r
groupInfo<-gsub("_\\w+", "", chiroptera$tip.label) #create grouping variable
```

```r
chiroptera <- groupOTU(chiroptera, groupInfo) #edit phylo object and add groupInfo to it
```

```r
head(fortify(chiroptera)) #does the underlying dataframe add gropInfo to it? yes, it does.
```

```r
ggtree(chiroptera, aes(color=group), layout='circular') + geom_tiplab(size=1, aes(angle=angle)) #color tree by groupInfo
```

```r
beast_file <- system.file("examples/MCC_FluA_H3.tree", package="ggtree") #working with phylogenetic data. load tree from BEAST
```

```r
beast_tree<-read.beast(beast_file)
```

```r
beast_tree
```

```r
ggtree(beast_tree,layout="circular")+geom_tiplab(aes(angle=angle))
```

```r
ggtree(beast_tree,layout="rectangular")+geom_tiplab()
```

```r
fortify(beast_tree) #look at underlying data. what are the different columns? 
```

```r
ggtree(beast_tree, branch.length = 'length') + theme_tree2() #scale tree by number of changes along branch (default)
```

```r
ggtree(beast_tree, branch.length = 'rate') +theme_tree2() # scale tree by rate of changes along branch 
```

```r
ggtree(beast_tree, branch.length = 'rate') + geom_range(range='rate_0.95_HPD', color='red', alpha=.6, size=2)
#add uncertainty measure to tree. HPD stands for high posterior density interval
```

```r
ggtree(beast_tree, branch.length = 'rate',layout="circular") + geom_range(range='rate_0.95_HPD', color='red', alpha=.6, size=2)
```
